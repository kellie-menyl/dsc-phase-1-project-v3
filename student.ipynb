#### NULL PERCENTAGE PER COLUMN

```python
# Displaying options to see all columns
pd.set_option('display.max.columns', 57)
# Calculate the percentage of null values in each column
null_percentage = (df.isnull().sum()/len(df)*100).round(2)
# converting it to a DataFrame for better display
null_percentage_df = null_percentage.to_frame(name='Null Percentage')
null_percentage_df
```

```python
# create a list of columns that have a null_percentage below 10
columns = [key for key, value in null_percentage.items() if value < 10]
# Drop rows with any missing values in the columns above
df.dropna(subset=columns,inplace= True)
```

## FILLING IN NULL VALUES

### INJURIES

```python
# Filling in the columns df['Total.Fatal.Injuries'], df['Total.Serious.Injuries'], df['Total.Minor.Injuries'] with 0
"""
Filling in the columns with 0 shows that data was either not collected or there are no injuries in those locations

"""
df['Total.Fatal.Injuries'] = df['Total.Fatal.Injuries'].fillna(0)
df['Total.Serious.Injuries'] = df['Total.Serious.Injuries'].fillna(0)
df['Total.Minor.Injuries'] = df['Total.Minor.Injuries'].fillna(0)
```

#### BROAD PHASE OF FLIGHT

```python
# Filling in the column df['Broad.phase.of.flight'] with 'Unknown'
df['Broad.phase.of.flight'].fillna('Unknown', inplace=True)
```

#### NORMALIZING THE MAKE COLUMN

```python
df['Make'] = df['Make'].str.upper()
```

#### AIRCRAFT CATEGORY

```python
# creating a function that uses the columns make and model 
def fill_nan_aircraft_category(df):
    df['Aircraft.Category'] = df.groupby('Make')['Aircraft.Category'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
    return df
fill_nan_aircraft_category(df)
```

#### DROPPING COLUMNS THAT I WILL NOT USE IN MY ANALYSIS

```python
# Dropping Latitude and Longitude columns
df.drop(columns=['Latitude', 'Longitude', 'Publication.Date', 'Airport.Name', 'Airport.Code', 'FAR.Description', 'Schedule', 'Air.carrier'], inplace=True)
```

## EXPLORING DATA USING VISUALIZATIONS

### 1. DISTRIBUTION OF ACCIDENTS OVER TIME

```python
# setting the size of the figure of the plot
plt.figure(figsize=(12, 6))
# Converting the 'Event.Date' column to datetime format
df['Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Date'].dt.year
# counting the number of accidents per year
accidents_count_per_year = df['Year'].value_counts().sort_index()

# creating a lineplot using seaborn to see the distribution
sns.lineplot(x=accidents_count_per_year.index, y=accidents_count_per_year.values)
plt.title('Number of Accidents count per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents count')
plt.grid(True)
plt.show()
```

From the visualization above we can clearly note that:
1. From the years before 1980 there were no airplanes
2. Airplanes were created from the year 1980
3. The number of accidents was recorded highest in that year
4. The number of accidents was reducing progressively due to the advancment in technology over the years
5. Improvement in the quality of airplanes created over the years


### 2. DISTRIBUTION OF ACCIDENTS BY AIRCRACT CATEGORY

```python
plt.figure(figsize=(10, 6))
# Creating a count plot using seaborn
sns.countplot(data=df, x='Aircraft.Category', order=df['Aircraft.Category'].value_counts().index, color = 'orange')
# Adding a title
plt.title('Distribution of Accidents by Aircraft Category')
# labeling the x-axis
plt.xlabel('Aircraft Category')
# labeling the y-axis
plt.ylabel('Number of Accidents')
# rotating the x-axis label for better readability
plt.xticks(rotation=90)
plt.yscale('log')

plt.show()
```

```python
# Number of each Aircraft category
df.groupby('Aircraft.Category')['Make'].count()
```

From the visualization above we clearly observe that:

3. The most used airplane category experienced the highest number of accidents
4. Airplane is the most used category

### 5. DISTRIBUTING OF INJURIES BY PHASE OF FLIGHT

```python
# Group by Broad.phase.of.flight and calculate totals
phase_of_flight_data = df.groupby('Broad.phase.of.flight').agg(
    total_incidents=('Event.Id', 'count'),
    total_fatal_injuries=('Total.Fatal.Injuries', 'sum'),
    total_serious_injuries=('Total.Serious.Injuries', 'sum'),
    total_minor_injuries=('Total.Minor.Injuries', 'sum')
).reset_index()
phase_of_flight_data
```

```python
# Setting the size of the figure
plt.figure(figsize=(14, 8))
# plotting first subplot
plt.subplot(2, 1, 1)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_incidents'], color='g')
plt.title('Total Incidents by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Incidents')
plt.xticks(rotation=45)
plt.grid(True)
# plotting second subplot
plt.subplot(2, 1, 2)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_fatal_injuries'], color='r', alpha=0.5, label='Fatal Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_serious_injuries'], color='orange', alpha=0.5, label='Serious Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_minor_injuries'], color='g', alpha=0.5, label='Minor Injuries')
plt.title('Total Injuries by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Injuries')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

```

Plotting two subplots:
1. The first subplot shows the total number of incidents by phase of flight.
